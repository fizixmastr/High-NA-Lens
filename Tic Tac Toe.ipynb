The next thing to handle is building the board. I know that the board has nine cells that will have to be filled. Of course it is possible to use a 3X3 array, however I feel that it is more pythonic to create a dicionary or places, and then call these the dictionary values into a printed board.

```python
#Create board dictionary
gameBoard = {'1': ' ' , '2': ' ' , '3': ' ' ,
            '4': ' ' , '5': ' ' , '6': ' ' ,
            '7': ' ' , '8': ' ' , '9': ' ' }
```

```python
#Will be printing the board a lot so set up a function
printBoard = lambda board:(
    print(' ' + board['1'] + '| ' + board['2'] + ' | ' + board['3'],
          '\n--+---+--\n', 
          board['4'] + '| ' + board['5'] + ' | ' + board['6'], 
          '\n--+---+--\n', 
          board['7'] + '| ' + board['8'] + ' | ' + board['9']))
```

This also seems like a good time to establish some way to clear the board as this will happen 2-4 time a game, and anytime the player wants to continue.

```python
#For clearing the board
board = []
for place in gameBoard:
    board.append(place)
```
